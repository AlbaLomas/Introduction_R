Gross domestic product (GDP) is a monetary measure of the market value of all the final goods and services produced in a specific time period. In spanish is called the PIB (Producto Interno Bruto).
The Human Development Index (HDI) is a statistic composite index of life expectancy, education (Literacy Rate, Gross Enrollment Ratio at different levels and Net Attendance Ratio), and per capita income indicators, which are used to rank countries into four tiers of human development.

Several online websites (among which the spanish wikipedia https://es.wikipedia.org/wiki/%C3%8Dndice_de_desarrollo_humano), have suggested that GDP and HDI are strongly correlated, meaning that the better the country's economy the better the human development of their citizens.
```r
x <- read.csv("GDP.csv") #x contains GDP data
y <- read.csv("HDI.csv") #y contains HDI data
```

- **Glue together the two data frames by columns (one next to the other) using cbind(). We want the province to match on the same rows, so you need first to sort the rows of the two data frames by Province, using order()**

```r
#Sort GDP by Province
xo <- x[order(x$Province), ]

#Sort HDI by Province
yo <- y[order(y$Province), ]
```

```r
new_dataframe <- cbind(xo,yo) #Creating a new data frame which combines GDP and HDI data
new_dataframe
```

- **The new data frame will contain two columns names "Province". remove one of the columns**

```r
df_new <- new_dataframe[,-4] #Removing the repeated column (Province), the data set without this repeated column is stores in a new data frame
df_new
```

- **Check that everything is fine with head() and tail(). The final data frame should have 4 columns: Province GDP.per.capita, CCAA and HDI. With each province matching its CCAA. (Fig, 1)***

```r
head(df_new) #This command shows that the first six lines are equal to the Fig 1. 
```

```r
tail(df_new) #This command shows the last six lines of the data frame
```

- **Create scatterplot of text using geom_point(). Plot the GDP.per.capita (x-axis) against the HDI (y-axis). Color the points by CCAA (use color=CCAA inside the aes() function of geom_point).**

```r
library(ggplot2)
ggplot(df_new, aes(x=GDP.per.capita, y=HDI, color=CCAA)) +
geom_point(size = 10) + theme(text = element_text(family = "Arial", size = 35, face = "bold"))
#theme element allows modifying the legend characteristics  
```

- **Do the same but use geom_text(). Specify as labels the names of the provinces, using aes(label=Province) inside geom_text. Color by CCAA. Also specify check_overlap=T so that the names do not overlap, if you want. (Fig, 2)**

```r
ggplot(df_new, aes(x=GDP.per.capita, y=HDI, color=CCAA)) +
geom_text(aes(label=Province), check_overlap=T, size = 13) + 
    theme(text = element_text(family = "Arial", size = 40, face = "bold"),legend.position = "bottom")

# Allow the plot to be bigger
options(repr.plot.width = 25, repr.plot.height = 20, repr.plot.res = 50)
```

- **Calculate the mean and standard deviation of GDP and HDI, by CCAA, using aggregate ().**

```r
#Calculating the mean of HDI by CCAA
mean_HDI <- aggregate(df_new$HDI, list(df_new$CCAA), mean)
#Fixing the column titles of HDI mean data frame
colnames(mean_HDI) <- c("CCAA", "Mean HDI")
```

```r
mean_HDI
```

```r
#Calculating the mean of GDP by CCAA
mean_GDP <- aggregate(df_new$GDP.per.capita, list(df_new$CCAA), mean)
#Fixing the column titles of GDP mean data frame
colnames(mean_GDP) <- c("CCAA", "Mean GDP")
```

```r
mean_GDP
```

```r
#Calculating the sd of GDP by CCAA
sd_GDP <- aggregate(df_new$GDP.per.capita, list(df_new$CCAA), sd)
#Fixing the column titles of GDP sd data frame
colnames(sd_GDP) <- c("CCAA", "SD GDP")
```

```r
sd_GDP
```

```r
#Calculating the sd of HDI by CCAA
sd_HDI <- aggregate(df_new$HDI, list(df_new$CCAA), sd)
#Fixing the column titles of HDI sd data frame
colnames(sd_HDI) <- c("CCAA", "SD HDI")
```

```r
sd_HDI
```

- **Create a new data frame where you put each CCAA with the corresponding mean GDP, mean HDI, stdev GDP and stdev HDI**

```r
#First, I created a data frame only for the HDI and GDP mean
df_mean<- cbind(mean_GDP, mean_HDI)
#this data frame has a repeated column (CCAA)
```

```r
#This line removes the repeated column
df_mean <- df_mean[,-3]
```

```r
df_mean
```

```r
#Then, I created a second data frame for sd of GDP and HDI 
df_SD<- cbind(sd_GDP, sd_HDI)
#this data frame has a repeated column (CCAA)
```

```r
#this line removes the repeated column 
df_SD <- df_SD[,-3]
```

```r
df_SD
```

```r
#Unify both dataframes
df_mean_sd <- cbind(df_mean, df_SD)
```

```r
#Removing the repeated column (CCAA)
df_mean_sd <- df_mean_sd[,-4]
```

```r
df_mean_sd
```

- **Add also vertical and horizontal error bars to express the variablity for each CCAA (with geom_errorbar and geom_errorbarh). For geom_errorbar() you have to specify the aestetic paramters x, y, ymin and ymax. For geom_errorbarh you have to specify the aestetic paramters x, y, xmin and xmax.**

```r
df_mean_sd
```

```r
#geom_errorbar : x = Mean GDP, ymin = Mean HDI - SD HDI, ymax = Mean HDI - SD HDI
#geom_errorbarh : y = Mean HDI, xmin = Mean GDP - SD GDP, xmax = Mean GDP + SD GDP

ggplot() + geom_point(data = df_new, aes(x=df_new[,2], y=df_new[,4], color=CCAA),  size=10) + theme(text = element_text(family = "Arial", size = 40, face = "bold")) + geom_point(data=df_mean_sd, x=df_mean_sd[,2], y=df_mean_sd[,3], size=10) +  
    geom_errorbar(data = df_mean_sd, aes(x = df_mean_sd[,2], ymin = df_mean_sd[,3] - df_mean_sd[,5], ymax = df_mean_sd[,3] + df_mean_sd[,5])) + 
    geom_errorbarh(data = df_mean_sd, aes(y = df_mean_sd[,3], xmin = df_mean_sd[,2] - df_mean_sd[,4], xmax = df_mean_sd[,2] + df_mean_sd[,4])) +
 labs(x = "GDP per capita",y = "HDI")
 
```

- **Finally, add the name of each CCAA on the plot with geom_label(). You can add the parameters  hjust=1, vjust=1 to move the labels a little bit so they don't overlap the points (Fig, 4)**

```r
ggplot() + geom_point(data = df_new, aes(x=df_new[,2], y=df_new[,4], color=CCAA),  size=15) + theme(text = element_text(family = "Arial", size = 45, face = "bold"),legend.position = "bottom") + 
    geom_point(data=df_mean_sd, x=df_mean_sd[,2], y=df_mean_sd[,3], size=10) +  
    geom_errorbar(data = df_mean_sd, aes(x = df_mean_sd[,2], ymin = df_mean_sd[,3] - df_mean_sd[,5], ymax = df_mean_sd[,3] + df_mean_sd[,5])) + 
    geom_errorbarh(data = df_mean_sd, aes(y = df_mean_sd[,3], xmin = df_mean_sd[,2] - df_mean_sd[,4], xmax = df_mean_sd[,2] + df_mean_sd[,4])) +
    geom_label(data=df_mean_sd, x=df_mean_sd[,2], y=df_mean_sd[,3], label = df_mean_sd[,1], size = 10, hjust=1, vjust=1) + labs(x = "GDP per capita",y = "HDI")
 
```

The correlation coefficient is a measure of association between two variables. Correlation values range from + 1 to - 1. 
- 1 is the perfect(directly proportional) correlation.
- 0 is the absence of correlation.
- -1 is the perfect (inversely proportional) correlation.

Let's calculate the correlation coefficient between the averages of GDP and HDI for each CCAA. 

```r
df_mean #This data frame contains the needed data
```

```r
x <- df_mean[,2] #Storing the GDP mean in x
y <- df_mean[,3] #Storing the HDI mean in y 
```

```r
cor(x,y) #calculating the correlation
```

This result says that GDP y HDP are significantly correlated

Using the data provided for the test, it can be seen that HDI and GDP are significantly correlated.
Interpreting the HDI formula, it can be concluded that an increase in GDP will in all cases imply an increase in HDI (if all other variables remain stable). However, it is not certain that an increase in the HDI implies an increase in GDP, since it could be caused by an increase in life expectancy or in the education index.
